   Handle missing data using Scikit-Learn: Using SimpleImputer()

```python
import pandas as pd
df=pd.read_csv('DATA FILES/weather_data.csv')
df
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN', strategy = "mean", axis = 0)
```

```python
df
```

```python
imputer = imputer.fit(df.iloc[:,1:3])
df.iloc[:,1:3] = imputer.transform(df.iloc[:,1:3])
df
```

```python
imputer = imputer.fit(df.iloc[:,1:3])
df.iloc[:,1:3] = imputer.transform(df.iloc[:,1:3])
df
```

# As Imputer object works only on the numerical values, for the categorical data we deal with modeÂ¶

```python
col = ['event']
df[col] = df[col].fillna(df.mode().iloc[0])
df
```

# Mean Imputation

```python
df.fillna(df.mean())
```

## Median Imputation

```python
df.fillna(df.median())
```
