```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.preprocessing import scale
```

```python
iris =  pd.read_csv('DATA FILES/iris.csv')
```

```python
iris
```

```python
iris.head()
```

```python
iris.shape
```

```python
iris.iloc[:,0:3].astype(int).head()
```

```python
iris.isnull().sum()
```

```python
X= iris.iloc[:,0:4]
y=iris.iloc[:,4]
```

```python
X=scale(X)
X
```

```python
X = pd.DataFrame(X)
X.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=1)
```

```python
y_test.head()
```

```python
model_knn = KNeighborsClassifier(n_neighbors=10, metric='euclidean')
model_knn.fit(X_train,y_train)
```

```python
model_knn
```

```python
y_predict = model_knn.predict(X_test)
np.mean(y_predict == y_test)
```

```python
confusion_matrix(y_predict,y_test)
```

```python
accuracy_score(y_predict,y_test)
```

```python
y_predict1 = model_knn.predict(X_test)
np.mean(y_predict1 == y_test)
```

```python
accuracy_score(y_test,y_predict1)
```

```python
confusion_matrix(y_test,y_predict1)
```

```python
np.mean(y_test == y_predict1)
```
